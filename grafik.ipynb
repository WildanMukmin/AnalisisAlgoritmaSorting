```python
import matplotlib.pyplot as plt

# Data for the sorting algorithms
n_values = [10, 100, 500, 1000, 10000]

# Bubble sort data
bubble_sort_random = [0, 0, 999, 6017, 562122]
bubble_sort_sorted = [0, 0, 1005, 1999, 239068]
bubble_sort_reversed = [0, 0, 1014, 6014, 624126]

# Insertion sort data
insertion_sort_random = [0, 0, 0, 2000, 167023]
insertion_sort_sorted = [0, 0, 0, 0, 0]
insertion_sort_reversed = [0, 0, 1000, 3998, 335075]

# Selection sort data
selection_sort_random = [0, 0, 1004, 3015, 226050]
selection_sort_sorted = [0, 0, 995, 2009, 227036]
selection_sort_reversed = [0, 0, 998, 1999, 232065]

# Merge sort data
merge_sort_random = [0, 0, 1010, 1000, 6990]
merge_sort_sorted = [0, 0, 0, 1002, 4000]
merge_sort_reversed = [0, 0, 0, 1005, 5001]

# Quick sort data
quick_sort_random = [0, 0, 0, 0, 1001]
quick_sort_sorted = [0, 0, 1006, 5001, 501154]
quick_sort_reversed = [0, 0, 1000, 3000, 307069]

# Plotting the data
plt.figure(figsize=(14, 10))

# Bubble Sort
plt.subplot(3, 2, 1)
plt.plot(n_values, bubble_sort_random, label='Random Data', marker='o')
plt.plot(n_values, bubble_sort_sorted, label='Sorted Data', marker='o')
plt.plot(n_values, bubble_sort_reversed, label='Reversed Data', marker='o')
plt.title('Bubble Sort')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time (µs)')
plt.legend()
plt.grid(True)

# Insertion Sort
plt.subplot(3, 2, 2)
plt.plot(n_values, insertion_sort_random, label='Random Data', marker='o')
plt.plot(n_values, insertion_sort_sorted, label='Sorted Data', marker='o')
plt.plot(n_values, insertion_sort_reversed, label='Reversed Data', marker='o')
plt.title('Insertion Sort')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time (µs)')
plt.legend()
plt.grid(True)

# Selection Sort
plt.subplot(3, 2, 3)
plt.plot(n_values, selection_sort_random, label='Random Data', marker='o')
plt.plot(n_values, selection_sort_sorted, label='Sorted Data', marker='o')
plt.plot(n_values, selection_sort_reversed, label='Reversed Data', marker='o')
plt.title('Selection Sort')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time (µs)')
plt.legend()
plt.grid(True)

# Merge Sort
plt.subplot(3, 2, 4)
plt.plot(n_values, merge_sort_random, label='Random Data', marker='o')
plt.plot(n_values, merge_sort_sorted, label='Sorted Data', marker='o')
plt.plot(n_values, merge_sort_reversed, label='Reversed Data', marker='o')
plt.title('Merge Sort')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time (µs)')
plt.legend()
plt.grid(True)

# Quick Sort
plt.subplot(3, 2, 5)
plt.plot(n_values, quick_sort_random, label='Random Data', marker='o')
plt.plot(n_values, quick_sort_sorted, label='Sorted Data', marker='o')
plt.plot(n_values, quick_sort_reversed, label='Reversed Data', marker='o')
plt.title('Quick Sort')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time (µs)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
